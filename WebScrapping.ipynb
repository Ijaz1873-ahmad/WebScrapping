```python
#import libraries
import time
from time import sleep
import pandas as pd
from selenium import webdriver
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

Scrape 3 pages of reviews for a single product listing using product URL.<br /> Note: Product must have more than 1 review page. The code must be generic and should work for every product listing URL, with proper checks in place.

```python
url = 'https://www.daraz.pk/products/luxury-6d-plating-soft-casing-silicone-square-frame-phone-case-for-samsung-galaxy-a32-4g-phone-case-shiny-bling-back-cover-coque-i228267660-s1448816145.html?spm=a2a0e.searchlist.list.1.4352287dh06hXB&search=1'

daraz_window = webdriver.Chrome(ChromeDriverManager().install())
daraz_window.maximize_window()

daraz_window.get(url)
sleep(7)

usernames = []
reviews = []
for i in range(3):
    print("Scraping Page", i+1)
    for i in range(5):
        sleep(2)
        xp = daraz_window.find_element(By.XPATH, '//*[@id="module_product_review"]/div/div/div[3]/div[1]/div['+str(i+1)+']/div[2]/span[1]')
        usernames.append(xp.text)
        rv = daraz_window.find_element(By.XPATH, '//*[@id="module_product_review"]/div/div/div[3]/div[1]/div['+str(i+1)+']/div[3]/div[1]')
        reviews.append(rv.text)

    next_button = daraz_window.find_element(By.XPATH,'//*[@id="module_product_review"]/div/div/div[3]/div[2]/div/button[2]')
    next_button.click()
    sleep(3)

daraz_window.close()
data = {
    'Name':usernames,
    'Review Text':reviews
}
df = pd.DataFrame(data)
df
```

Scrape 80 items for a keyword search e.g., 'mobile covers'<br/>Note: Must use "click()" function of webElement, URL changing for pages is not allowed. The code must be generic and should work for every keyword, with proper checks in place. Bonus for also scraping country and Daraz Mall availability.

```python
query = 'mobile covers'
query = query.replace(' ','%20')
url = f'https://www.daraz.pk/catalog/?from=input&q={query}'
print(url)
```

```python
daraz_window = webdriver.Chrome(ChromeDriverManager().install())
daraz_window.maximize_window()
daraz_window.get(url)
sleep(3)

product_name = []
rate_List = []
stars = []
for i in range(2):
    print("Scrapping page", i+1)
    for i in range(40):
        product = daraz_window.find_element(By.XPATH, '//*[@id="root"]/div/div[2]/div/div/div[1]/div[2]/div['+str(i+1)+']/div/div/div[2]/div[2]/a')
        product_name.append(product.text)

        price = daraz_window.find_element(By.XPATH,'//*[@id="root"]/div/div[2]/div/div/div[1]/div[2]/div['+str(i+1)+']/div/div/div[2]/div[3]/span')
        rate_List.append(price.text)
    
    if i == 0:
        next_page = daraz_window.find_element(By.XPATH,'//*[@id="root"]/div/div[2]/div/div/div[1]/div[3]/div/ul/li[9]/a')
        next_page.click()
        sleep(3)

daraz_window.close()
```

```python
daraz_window = webdriver.Chrome(ChromeDriverManager().install())
daraz_window.maximize_window()
daraz_window.get(url)

sleep(3)

xp = '//*[@id="module_product_review"]/div/div/div[1]/div[2]/div/div/div[1]/div[3]'
xp = '//*[@id="module_product_review"]/div/div/div[1]/div[2]/div/div/div[1]/div[3]'
stars = []
ratings = []
for i in range(2):
    for j in range(40):
        button = daraz_window.find_element(By.XPATH, '//*[@id="root"]/div/div[2]/div/div/div[1]/div[2]/div['+str(j+1)+']/div/div/div[2]/div[2]/a')
        button.click()
        try:
            sleep(2)
            stars.append(daraz_window.find_element(By.XPATH,'//*[@id="module_product_review"]/div/div/div[1]/div[2]/div/div/div[1]/div[1]').text)
        except NoSuchElementException:
            stars.append("No-Rating")
        
        try:
            sleep(2)
            ratings.append(daraz_window.find_element(By.XPATH, '//*[@id="module_product_review"]/div/div/div[1]/div[2]/div/div/div[1]/div[3]').text)
        except NoSuchElementException:
            ratings.append("No-Review")
        
        daraz_window.back()
        sleep(1)
        
    if i == 0:
        next_page = daraz_window.find_element(By.XPATH,'//*[@id="root"]/div/div[2]/div/div/div[1]/div[3]/div/ul/li[9]/a')
        next_page.click()
        sleep(3)

daraz_window.close()
```

```python
for i in range(len(stars)):
    stars[i] = stars[i].replace("/5", "")
    
for i in range(len(ratings)):
    ratings[i] = ratings[i].replace("Ratings", "")
    
for i in range(len(rate_List)):
    rate_List[i] = rate_List[i].replace("Rs.", "")
```

```python
data = {
    'Name':product_name,
    'Price':rate_List,
    'Stars':stars,
    'Reviews': ratings
}
df = pd.DataFrame(data)
df.head(5)
```
